```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
UFO_file = "Resources/UFO-scrubbed.csv"
UFO_df = pd.read_csv(UFO_file)
UFO_df.head()
```

```python
military_file = "Resources/military-bases.csv"
military_df = pd.read_csv(military_file)
military_df.head()
```

```python
reduced_UFO_df = UFO_df.loc[:, ['datetime', 'city', 'state', 'country', 'shape','date posted', 'latitude',
       'longitude ']]
reduced_UFO_df.head()
```

```python
columns = ['datetime', 'city', 'state', 'country', 'shape','date posted', 'latitude',
       'longitude ']

reduced_UFO_df = reduced_UFO_df.loc[reduced_UFO_df["country"] == "us", columns]

reduced_UFO_df.head()
```

```python
reduced_UFO_df.rename(index={'us': 'United States'})
reduced_UFO_df.head()
```

```python
clean_UFO_df= reduced_UFO_df.dropna(how="any")
clean_UFO_df.head()
```

```python
clean_UFO_df = clean_UFO_df.drop_duplicates(subset=None, keep='first', inplace=False)
clean_UFO_df.head()
```

```python
reduced_military_df = military_df.loc[:, ['Column1.datasetid','Column1.fields.oper_stat', 'Column1.fields.state_terr', 'Column1.fields.site_name', 'Column1.fields.component', 'Column1.fields.geo_point_2d',
       'Column1.fields.country', 'Column1.geometry.coordinates']]

reduced_military_df.head()
```

```python
renamed_military_df = reduced_military_df.rename(columns={'Column1.datasetid':'Data Id', 'Column1.fields.oper_stat': 'Base Status',
       'Column1.fields.state_terr':'state', 'Column1.fields.site_name':'Base Name',
       'Column1.fields.component':'Base Type', 'Column1.fields.country':'country',
       'Column1.fields.geo_point_2d':'latitude', 'Column1.geometry.coordinates':'longitude',})

renamed_military_df.head()
```

```python
columns = ['Data Id', 'Base Status', 'state', 'Base Name', 'Base Type', 'latitude',
       'country', 'longitude']

reduced_military_df = renamed_military_df.loc[renamed_military_df["country"] == "United States",  columns]
reduced_military_df.head()
```

```python
clean_military_df = reduced_military_df.drop_duplicates(subset=['Base Name'], keep='first', inplace=False)
clean_military_df
```

```python
clean_military_df = clean_military_df.dropna(how="any")
clean_military_df
```
