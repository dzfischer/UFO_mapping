```python
military_file = "Resources/military-bases.csv"
military_df = pd.read_csv(military_file)
military_df.head()
```

```python
reduced_UFO_df = UFO_df.loc[:, ['datetime', 'city', 'state', 'country', 'shape','date posted', 'latitude',
       'longitude ']]
reduced_UFO_df.head()
```

```python
columns = ['datetime', 'city', 'state', 'country', 'shape','date posted', 'latitude',
       'longitude ']

reduced_UFO_df = reduced_UFO_df.loc[reduced_UFO_df["country"] == "us", columns]

reduced_UFO_df.head()
```

```python
clean_UFO_df = clean_UFO_df.drop_duplicates(subset=None, keep='first', inplace=False)
clean_UFO_df.head()
```

```python
reduced_military_df = military_df.loc[:, ['Column1.datasetid','Column1.fields.oper_stat', 'Column1.fields.state_terr', 'Column1.fields.site_name', 'Column1.fields.component', 'Column1.fields.geo_point_2d',
       'Column1.fields.country', 'Column1.geometry.coordinates']]

reduced_military_df.head()
```

```python
columns = ['Base Status', 'state', 'Base Name', 'Base Type', 'latitude',
       'country', 'longitude']

reduced_military_df = renamed_military_df.loc[renamed_military_df["country"] == "United States",  columns]
reduced_military_df.head()
```

```python
clean_military_df = reduced_military_df.drop_duplicates(subset=['Base Name'], keep='first', inplace=False)
# clean_military_df
```

```python
clean_military_df = clean_military_df.dropna(how="any")
clean_military_df.head()
# clean_military_df
```

```python
clean_military_df.reset_index(inplace = True, drop = True)
clean_military_df.head()
# clean_military_df
```

```python
print(f"UFO table: {len(clean_UFO_df)}\nMilitary table: {len(clean_military_df)}")
clean_UFO_df.reset_index(inplace = True, drop = True)
clean_UFO_df.head()
```

```python
final_UFO_df = clean_UFO_df
final_UFO_df['nearest base'] = ""
final_UFO_df['nearest base (mi)'] = ""
final_UFO_df['nearest airbase'] = ""
final_UFO_df['nearest airbase (mi)'] = ""
final_UFO_df.head()
```

```python
# USE 'pip install geopy'
from geopy import distance
# distance.distance(latitude, longitude).miles
```

```python
# WARNING: THIS CODE TAKES 1HR+ TO RUN, BE PATIENT (or optimize it, I don't care, I'm not a cop)

airbase_dist_list = []
airbase_name_list = []
base_dist_list = []
base_name_list = []

# big loop through list of UFO sightings
for i in range(len(clean_UFO_df)):
    
    # record the location of this UFO sighting, have default closest sightings at v large range
    ufo_loc = (clean_UFO_df.loc[i,'latitude'], clean_UFO_df.loc[i,'longitude'])
    all_closest = 1800
    air_closest = 1800
    
    for j in range(len(clean_military_df)):
        # record the location of this base
        mil_loc = (clean_military_df.loc[j,'latitude'], clean_military_df.loc[j,'longitude'])
        
        # for each military base (for each UFO sighting), use geopy to find the approximate distance
        # note it's slow af, but... well, so be it.
        base_dist = distance.distance(mil_loc, ufo_loc).miles
    
        # if this is the closest base thus far, remember the distance and name
        if base_dist < all_closest:
            all_closest = base_dist
            all_base_name = clean_military_df.loc[j, 'Base Name']
            
        # if it's the closest air, also make a note
        if ((base_dist < air_closest) and (clean_military_df.loc[j,'Base Type'] == ( 'AF Active' or 'AF Guard' ))):
            air_closest = base_dist
            air_base_name = clean_military_df.loc[j, 'Base Name']           

    # once we've found the best bases for this sighting, remember it
    airbase_dist_list.append(air_closest)
    airbase_name_list.append(air_base_name)
    base_dist_list.append(all_closest)
    base_name_list.append(all_base_name)
    
    # just to give a readout that something's working
    if i%63 == 0:
        print(f"Sighting {i} in {clean_UFO_df.loc[i,'city']}: Closest: {all_base_name} : {all_closest:.2f} :"
              f": Closest air : {air_base_name} : {air_closest:.2f}.")

final_UFO_df['nearest base'] = base_name_list
final_UFO_df['nearest base (mi)'] = base_dist_list
final_UFO_df['nearest airbase'] = airbase_name_list
final_UFO_df['nearest airbase (mi)'] = airbase_dist_list
```
