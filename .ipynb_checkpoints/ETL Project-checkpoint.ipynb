```python
def dist(x1, x2, y1, y2):
    # latitudes (x) go from 0 to +- 90 as opposed to longitude's +- 180, Factor of 2 required to normalize to longitude
    distance = (4*(x2-x1)**2 + (y2-y1)**2)**(1/2)
    return distance

def findmin(lat, lon):
    close_series = dist(lat, clean_military_df["latitude"], lon, clean_military_df["longitude"])
#     print(close_series)
    closest = min(close_series)
    item = pd.Index(close_series).get_loc(closest)
#     print(item)
    return closest, item

# test = findmin(29, -97)
# print(test)
# clean_military_df.loc[[606]]

test = findmin(clean_UFO_df['latitude'], clean_UFO_df['longitude'])


# final_UFO_df = clean_UFO_df
# final_UFO_df['nearest base','distance'] = findmin(clean_UFO_df['latitude'], clean_UFO_df['longitude'])
```
